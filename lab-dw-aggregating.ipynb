```python
import pandas as pd

df_marketing = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv")

pd.set_option("display.max_columns",None)
```

```python
df_marketing.head(3)
```

```python
# 1. Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000 and have a response of "Yes" to the last marketing campaign.
new_df = df_marketing[(df_marketing["Total Claim Amount"] > 1000) & (df_marketing["Response"] == "Yes")].reset_index()

new_df
```

```python
#2. Using the original Dataframe, analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign.
# Write your conclusions.

new_df_2 = pd.pivot_table(df_marketing, index="Gender", columns=["Policy Type", "Response"], values=["Total Claim Amount"], aggfunc="mean")
```

```python
new_df_2

#Tanto hombres como mujeres que han dicho que sí en la encuesta han reclamado una cantidad muy similar.
#Aunque las mujeres que tienen póliza Corporativa o Especial han solicitado <1% más que los hombres, mientras que en el caso de las pólizas personales es al revés, son los hombres los que han reclamado un 1% más.
```

```python
# Analyze the total number of customers who have policies in each state

customers_per_state = df_marketing[["State", "Customer"]].groupby("State").count()

filtered_customers_per_state = customers_per_state[customers_per_state.Customer > 500]

filtered_customers_per_state
```

```python
#and then filter the results to only include states where there are more than 500 customers.

filtered_customers_per_state = customers_per_state[customers_per_state.Customer > 500]

filtered_customers_per_state
```

```python
#Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

median_customer_lifetime_value_per_education_and_gender = pd.pivot_table(df_marketing, index="Education", columns=["Gender"], values=["Customer Lifetime Value"], aggfunc="mean")
max_customer_lifetime_value_per_education_and_gender = pd.pivot_table(df_marketing, index="Education", columns=["Gender"], values=["Customer Lifetime Value"], aggfunc="max")
min_customer_lifetime_value_per_education_and_gender = pd.pivot_table(df_marketing, index="Education", columns=["Gender"], values=["Customer Lifetime Value"], aggfunc="min")
```

```python
median_customer_lifetime_value_per_education_and_gender

#Las mujeres que más tiempo contratan los servicios de la aseguradora son las que tienen menor nivel educativo. Mientras que en el caso de los hombres hay prácticamente un empate entre los que tienen menor nivel educativo y los que tienen máster.
#Las mujeres y hombres que menos tiempo contratan los servicios de la aseguradora son las/los que tienen mayor nivel educativo (doctorado).
```
